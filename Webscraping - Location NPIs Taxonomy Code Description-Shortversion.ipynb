```python
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--ignore-ssl-errors')
options.add_argument("--disable-notifications")
driver = webdriver.Chrome(driver_path, chrome_options=options)
```

```python
url = 'https://npiregistry.cms.hhs.gov/'
driver.get(url)
```

```python
for i in range(len(NPI_list)): 
    try:
        t = pd.Timestamp.now()
        x = NPI_list.loc[i,"Key_NPI" ].astype(str)
        print(x)
        what_search_box = driver.find_element_by_xpath('//*[@id="id_number"]')
        what_search_box.clear()
        what_search_box.send_keys(x)
        what_search_box.send_keys(Keys.ENTER)

        try:
            NPI_list.loc[i,"Name"] = driver.find_element_by_xpath ('/html/body/div[2]/div[2]/div/table/tbody/tr/td[2]').text
            NPI_list.loc[i,"PrimaryPracticeAddress"] = driver.find_element_by_xpath ('/html/body/div[2]/div[2]/div/table/tbody/tr/td[4]').text
#             NPI_list.loc[i,"PrimaryStateandzip"]     = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/table/tbody/tr/td[4]/text()[2]').text
            NPI_list.loc[i,"Primary Taxonomy"]       = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/table/tbody/tr/td[6]/div').text
#             time.sleep(1)    
            print(i,pd.Timestamp.now() - t, NPI_list.loc[i,['Name','PrimaryPracticeAddress','Primary Taxonomy' ]])
        except: 
            pass

        newSearch = driver.find_element_by_xpath('/html/body/div[2]/div[3]/div[2]/div/form/button')
        newSearch.send_keys(Keys.ENTER)

    except:
        pass
```
